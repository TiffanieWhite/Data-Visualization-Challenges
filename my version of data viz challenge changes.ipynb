```python
import pandas as pd
import matplotlib.pyplot as plt

coffee_df = pd.read_csv("fake_coffee_chain_sales.csv")
coffee_df.head()
```

```python
plt.figure(figsize=(8,5))
subset = coffee_df.groupby("Year")["Sales"].sum().reset_index()

# Wrong variable for title, truncated y-axis, misleading color & framing
plt.plot(subset["Year"], subset["Sales"], color="green", linewidth=4, marker="o", linestyle=':')
#plt.ylim(subset["Sales"].min() * 0.98, subset["Sales"].min() * 1.02)  # extreme truncation

plt.title("Coffee Sales Profits from 2020-2024", fontsize=16, fontweight="bold", color="blue")
plt.xticks(fontsize=8, rotation=45)
plt.xticks(range(2020, 2025))
plt.yticks(fontsize=8)

plt.xlabel('Years(Sales)')
plt.ylabel('Percentage of Sales')
plt.tight_layout()


ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_ylim(ymin=0)
plt.show()
```

1.The first thing I did was to adjust the font size because the wording was all running in together. 
2. I then rotated the x axis 45 degrees. 
3. The video about plot ethics informed us to try not to use red as it is not a good color for those with accessibility issues and it can show alarm/bias. 
4. I changed the title from Dark Red and renamed to reflect the coffee sales years. 
5. I labeled the x and y sides. 
6. I removed the spines on the top and the right. 
7. I set the y axis to start at "0". 
8. I changed the linestyle of the plot
9. I changed the x ticks range to make the numbers integers instead of floats. 









```python
plt.figure(figsize=(10,6))
avg_sales = coffee_df.groupby("Region")["Sales"].mean().reset_index()

colors = ["#443168","#081c52","#615eeb","#3807c0","#4e4cf0","#1c133f"]

plt.bar(avg_sales["Region"], avg_sales["Sales"], 
        #color=colors, edgecolor="black", linewidth=3, hatch="/", alpha=0.8)
        color=colors, edgecolor="black", linewidth=3, alpha=0.8)
plt.title("Coffee Sales Totals by Region", fontsize=20, fontweight="bold", color="black", backgroundcolor="white")
#plt.grid(True, which='both', color='black', linestyle='-', linewidth=1.5)
plt.grid(True, axis='y', color='black', linestyle='-', linewidth=1.5)

plt.xlabel("The Four International Regions", fontsize=16, fontweight="bold", color="darkblue")
plt.ylabel("Average Amount of Coffee Sold($)", fontsize=16, fontweight="bold", color="darkblue")
#plt.xticks(rotation=45, fontsize=14, fontweight="bold", color="orange")
plt.xticks(fontsize=12, fontweight="bold", color="black")
#plt.yticks(fontsize=14, fontweight="bold", color="orange")
plt.yticks(rotation= 70, fontsize=12, fontweight="bold", color="black")

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)





plt.show()
```

1. I changed the title and removed the exclamation marks. They were drawing attention to the title instead of the data. 
2. I changed the title from purple to black. 
3. I changed the title background from yellow to clear. 
4. I changed the title on the y axis to the amount of coffee sold. 
5. I changed the color scheme to similar colors so as not to be drawn to just one region. 
6. I changed the title of the x axis
7. I removed the rotation of the column headers for the x axis, changed the font size, and changed the color to black. 
8. I added rotation to the y axis, changed the font size, and changed the color to black. 
9. I removed the top and right spines 
10. I removed the hatch "/". I didn't see the need for the extra lines. 
11. I removed the full grid and just placed a grid for the y axis
